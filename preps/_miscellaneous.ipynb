```python
# Smallest Value of the Rearranged Number

# You are given an integer num. Rearrange the digits of num such that its value is minimized and it does not contain any leading zeros.
# Return the rearranged number with minimal value.
# Note that the sign of the number does not change after rearranging the digits.

'''
# interface
Args:
    num: number like . not 0. (-12, 15)
def min_num_after_rearrange(num) -> return number


# example
20

->
20
x 02
----> return 20.



- positive  (just find the smallest)
12
-> return 12
1340
-> 0134 xxxxx
-> 1034 xxxxx


- negative (find the largest, and return negative of it.)
-12
-> return -21

# algorithm

if pos
    sort f in ascending order
    if digits[0] == 0 and 2 <= len(digits) -> swap first 0 and second next non 0.
    return concatanated num
else
    sort digits in descending order
    return - concatanated num

assume N = num digits,
time  = O(NlogN + N) = O(NlogN)
space = O(N)
'''

# 2001
def min_num_after_rearrange(num):
    digits = [int(d) for d in str(abs(num))] # [2,0,0,1]

    if 0 < num:
        digits.sort() # [0,0,1,2]
        digits = avoid_trailing_zero(digits) # [1,0,0,2]
        return int("".join([str(d) for d in digits]))
    else:
        digits.sort(reverse=True)
        return - int("".join([str(d) for d in digits]))


def avoid_trailing_zero(digits): # [1,0,0,2]
    if digits[0] != 0:
        return digits
    
    for i in range(1, len(digits)): # 1->2
        if digits[i] != 0:
            break
    
    digits[0], digits[i] = digits[i], digits[0]
    return digits


# test

assert min_num_after_rearrange(1) == 1
assert min_num_after_rearrange(-1) == -1

assert min_num_after_rearrange(312) == 123
assert min_num_after_rearrange(-312) == -321

assert min_num_after_rearrange(3102) == 1023
assert min_num_after_rearrange(-3102) == -3210

assert min_num_after_rearrange(301002) == 100023
assert min_num_after_rearrange(-301002) == -321000
```

```python
# Sales Path (from Pramp)

class Node:
  def __init__(self, cost):
    self.cost = cost
    self.children = []
    self.parent = None

def get_cheapest_cost(root):
    if root is None:
        return None
    return get_cheapest_cost_helper(root)

# Don't use shared var if I can do without it.
# In many cases, the code is simpler.
def get_cheapest_cost_helper(root):
  is_leaf = len(root.children) == 0
  if is_leaf:
    return root.cost

  min_cost = float('inf')
  for child in root.children:
    cost = root.cost + get_cheapest_cost_helper(child)
    min_cost = min(min_cost, cost)
  
  return min_cost
    

# For backtracking, I have to keep the record of cheapest cost path.
# For this purpose, I will use global_min_cost and global_min_cost_path.
def get_cheapest_cost_path(root):
    global_min_cost = float('inf')
    global_min_cost_path = []

    # Side effect: updates global_min_cost and global_min_cost_path
    def traverse(root, nodes_so_far, cost_so_far):
        if root is None:
            return

        nodes_so_far.append(root.cost)
        cost_so_far += root.cost

        is_leaf = len(root.children) == 0
        if is_leaf:
            nonlocal global_min_cost, global_min_cost_path
            if cost_so_far < global_min_cost:

                global_min_cost = cost_so_far
                global_min_cost_path = nodes_so_far.copy()

        for child in root.children:
            traverse(child, nodes_so_far, cost_so_far)

        # because nodes_so_far is shared among function calls.
        nodes_so_far.pop()
            
    traverse(root, [], 0)
    return global_min_cost_path
        

# test
tree = Node(1)
tree.children = [
  Node(2), Node(3)
]
tree.children[0].children = [
  Node(2)
]

assert get_cheapest_cost(tree) == 4
print(get_cheapest_cost_path(tree))
print(get_cheapest_cost_path(None))
```
