```python
# Insert Intervals

def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
    merged = []
    i1 = 0 # to be inserted next
    new_already_inserted = False

    while i1 <= len(intervals) - 1 or not new_already_inserted:
        if i1 == len(intervals):
            chosen = newInterval
            new_already_inserted = True
        elif new_already_inserted:
            chosen = intervals[i1]
            i1 += 1
        elif intervals[i1][0] <= newInterval[0]:
            chosen = intervals[i1]
            i1 += 1
        else:
            chosen = newInterval
            new_already_inserted = True
        # print(chosen)

        # append chosen into merged array
        if len(merged) == 0 or merged[-1][1] < chosen[0]:
            merged.append(chosen)
        else:
            merged[-1][1] = max(merged[-1][1], chosen[1])
    
    return merged
```

```python
# Merge 2 sorted interval arrays

# ref. https://leetcode.com/discuss/interview-question/883800/facebook-phone-merge-sorted-arrays-merge-two-sorted-interval-lists
#      https://leetcode.com/discuss/interview-question/124616/

# Q. Given two Lists of sorted intervals merge the overlapping items
# 
# List[int[]] listA = {[1,3], [4,5], [7,10], [16, 17]};
# List[int[]] listB = {[2,3], [4,8], [11,15]};
# 
# Expected result:
# {[1,3], [4, 10], [11, 15], [16, 17]}

'''

1-----4 
  2-------6
             7---10
             ^i1

    3---5
          6----------12
                           ^i2


1----5
       6---7

time  = O(N1 + N2)
space = O(1) aside from merged intervals to be returned.
'''


# (1,3),(3,5)
# (2,4)
# -> return (1,5)
def merge_sorted_intervals(arr1, arr2):
    merged_intervals = [] # (1,3) -> (1, 4) -> (1, 5)
    i1 = 0 # 0->1->2
    i2 = 0 # 0->1

    while i1 <= len(arr1)-1 or i2 <= len(arr2)-1: #  i1 <= 1 or i2 <= 0
        # all the intervals in arr1 are traversed already.
        if i1 == len(arr1): # i1 == 2
            next_interval = arr2[i2]
            i2 += 1
        # all the intervals in arr2 are traversed already.
        elif i2 == len(arr2): # i2 == 1
            next_interval = arr1[i1]   ######## (3,5)
            i1 += 1
        # pick up interval from arr1
        elif arr1[i1][0] <= arr2[i2][0]: # 3 < 2
            next_interval = arr1[i1]
            i1 += 1
        # pick up interval from arr2
        else:
            next_interval = arr2[i2]
            i2 += 1

        # append next_interval to merged_intervals
        if len(merged_intervals) == 0 or merged_intervals[-1][1] < next_interval[0]: # 4 < 3
            merged_intervals.append(next_interval)
        else:
            merged_intervals[-1][1] = max(merged_intervals[-1][1], next_interval[1])

    return merged_intervals
    

assert merge_sorted_intervals(
    [[1,3], [4,5], [7,10], [16, 17]],
    [[2,3], [4,8], [11,15]],
) == [[1,3], [4, 10], [11, 15], [16, 17]]
    
assert merge_sorted_intervals(
    [[1,3], [3,5]],
    []
) == [[1,5]]

assert merge_sorted_intervals([], []) == []
```

```python
# Meeting Rooms

# Given an array of meeting time intervals where intervals[i] = [starti, endi], determine if a person could attend all meetings.


'''
len(interval) == 0 -> return True.
[1,5][5,7] -> attend both of them.

def can_attend_all(intervals): -> return bool

[[1,5],[2,3]]
1--------5
   2--3         -> return false

[[1,5],[4,7]]
1-----5
   4------7     -> return false

[[1,5],[5,7]]
1---5
    5---7       -> return True


1-----5
   2--------7
          6-----9
                  10----12  

Sort intervals
for each interval,
    compare with prev interval -> return False if overlaps
return True

time  = NlogN
space = N        (for sorting)
'''

def can_attend_all(intervals):# [1,5],[7,10],[4,6]
    if len(intervals) == 0:
        return True

    intervals.sort() # [1,5],[4,6],[7,10],
    for i in range(1, len(intervals)): # *1,2
        prev_interval_end = intervals[i-1][1] #5
        curr_interval_start = intervals[i][0] # 4

        if curr_interval_start < prev_interval_end:
            return False
    return True

assert can_attend_all([[1,5],[7,10],[4,6]]) is False # False
assert can_attend_all([[1,5],[2,3]]) is False
assert can_attend_all([[1,5],[4,7]]) is False
assert can_attend_all([[1,5],[5,7]]) is True

assert can_attend_all([]) is True
```

```python
# Meeting Rooms II

# Given an array of meeting time intervals intervals where intervals[i] = [starti, endi],
# return the minimum number of conference rooms required.


'''
# Interface
if len(intervals) == 0 -> return 0
def min_conference_rooms(intervals): -> return int

# Examples
1---------5
     3--4
   2------------9

-> return 3.

# Approach
+1 +1  +1  -1   -1 -1
1---------------5
    2--------------9
        3--4

for each interval:
    append start event and end event to events
sort events
traverse events and track max_num_rooms.

# time  = NlogN
# space = N + N = N
'''
# Impl
def min_conference_rooms(intervals): # [1,3],[3,5]
    events = [] # (1,1),(3,-1),(3,1),(5,-1)
    for start, end in intervals:
        events.append((start, +1)) 
        events.append((end, -1))
    events.sort() # (1,1),(3,-1),(3,1),(5,-1)

    max_dup_events = 0 # 0->1->....
    curr_num_events = 0 # 0->1->0->1->0
    for _timestamp, diff_num in events: # 1, -1, 1, -1
        curr_num_events += diff_num
        max_dup_events = max(max_dup_events, curr_num_events)
    return max_dup_events

# Test

assert min_conference_rooms([[1,3],[3,5]]) == 1
assert min_conference_rooms([[1,3],[2,5]]) == 2
assert min_conference_rooms([[1,3],[0,5]]) == 2
assert min_conference_rooms([]) == 0
```

```python
# Non-overlapping Intervals
# Given an array of intervals intervals where intervals[i] = [starti, endi],
# return the minimum number of intervals you need to remove to make the rest of the intervals non-overlapping.

'''
# Interface
Args:
    intervals: an array of intervals
def min_num_to_remove_for_no_overlaping(intervals) -> return int.
if num(intervals) == 0 -> return 0

# Example

intervals: e.g.
[
    [1, 4],
    [3, 7],
    [8, 11],
    [9, 10],
]

1----4
   3----7
           8--------11
              9--10


# Algorithm



  1 2  3  
1-------6*
   3---------8           xxxx       
      4---------9        xxxx
        6----------10    xxxx
   
           7-------10    xxxx
                9------12
                          13-------15
                               14-------16


1--------------------10 xxxxx
   2--------------9     xxxxx
      3--------7*
         4-----------------12   xxxxxxx
                  9--------------------------20

sort intervals

del_cnt = 0
curr_earliest_end

for each interval:
    if prev_end < start:
        prev_end = end
        continue

    del_cnt += 1
    update prev_end to be min(prev_end, curr_end)
return del_cnt
'''

# Impl
def min_num_to_remove_for_no_overlaping(intervals):
    intervals.sort()

    del_cnt = 0 # 1->2->3
    prev_end = - float("inf") # 10->20

    for start, end in intervals:  # *[1,10], [2,9], [3,7], [4,12], [9,20]
        is_duplicated_with_prev = start < prev_end
        if is_duplicated_with_prev:
            del_cnt += 1
            prev_end = min(prev_end, end)
        else:
            prev_end = end
    return del_cnt

# Test
assert min_num_to_remove_for_no_overlaping([[1,10], [2,9], [3,7], [4,12], [9,20]]) == 3

assert min_num_to_remove_for_no_overlaping([[1,5], [6, 10]]) == 0
assert min_num_to_remove_for_no_overlaping([[1,5], [4, 10]]) == 1
assert min_num_to_remove_for_no_overlaping([[1,5], [2, 4]])  == 1
assert min_num_to_remove_for_no_overlaping([]) == 0
```

```python
# You are given two lists of closed intervals, firstList and secondList,
#                                              where firstList[i] = [starti, endi] and secondList[j] = [startj, endj].
# Each list of intervals is pairwise disjoint and in sorted order.

# Return the intersection of these two interval lists.
# A closed interval [a, b] (with a <= b) denotes the set of real numbers x with a <= x <= b.

# The intersection of two closed intervals is a set of real numbers that are either empty or represented as a closed interval.
# For example, the intersection of [1, 3] and [2, 4] is [2, 3].

''''
# interface

args:
    list1, list2 can be empty. -> return []
def intesection(list1, list2): -> return list of intervals e.g. [[1,2], [3,4]]

-- [1,2], [2,3] -> [2] is the intersection.

# example

1----2   3-----4   5---------7 
     2------------------6    7----8

-> return [[2,2],[3,4]]

1----2           4--------------7
         3-------4    5----6

-> return [4,4],[5,6]

# algorithm

time  = O(N1 + N2)
space = O(1)

Use 2 pointers, p1 and p2.

p1 = 0
p2 = 0

while p1 < len(l1) and p2 < len(l2):
    get the intersection.
    compare i1 and i2. if i-x-end < i-y-end, move p-x


1------------------5   6--7    8----10 11-----12   13----14

   2-----3   4-----------------8                   13-----14

'''

# Impl
def intersection(l1, l2):
    if len(l1) == 0 or len(l2) == 0:
        return []

    p1 = 0
    p2 = 0
    intersections = [] #(1,2),(3,5),

    # l1 = [1,2], *[3,6]
    # l2 = [1,5], *[7,10]

    while p1 <= len(l1) - 1 and p2 <= len(l2) - 1: # p1<=1 and p2<=1
        curr_intersection = intersection_of(l1[p1], l2[p2])
        if curr_intersection is not None:
            intersections.append(curr_intersection)

        end1 = l1[p1][1] # 6
        end2 = l2[p2][1] # 10
        
        if end1 < end2:
            p1 += 1
        else:
            p2 += 1

    return intersections
        
def intersection_of(i1, i2):
    start1, end1 = i1
    start2, end2 = i2

    if end1 < start2:
        return None
    if end2 < start1:
        return None

    return [max(start1, start2), min(end1, end2)]

# Test
print(intersection(
    [[1,2],[3,6]],
    [[1,5],[7,10]],
))
assert intersection(
    [[1,2],[3,6]],
    [[1,5],[7,10]],
) == [[1,2], [3,5]]

assert intersection(
    [[1,2]],
    [[1,3]],
) == [[1,2]]

assert intersection(
    [[2,4]],
    [[1,3]],
) == [[2,3]]

assert intersection(
    [[2,4]],
    [[5,10]],
) == []

assert intersection([],[[1,2]]) == []
assert intersection([],[]) == []
```

```python
# A set of real numbers can be represented as the union of several disjoint intervals,
# where each interval is in the form [a, b).

# A real number x is in the set if one of its intervals [a, b) contains x (i.e. a <= x < b).

# You are given a sorted list of disjoint intervals intervals representing a set of real numbers as described above,
# where intervals[i] = [ai, bi] represents the interval [ai, bi).

# You are also given another interval toBeRemoved.

# Return the set of real numbers with the interval toBeRemoved removed from intervals.
# In other words, return the set of real numbers such that every x in the set is in intervals but not in toBeRemoved.
# Your answer should be a sorted list of disjoint intervals as described above.


'''
# Interface
Args:
    intervals: sorted list of disjoint intervals, represents the interval [ai, bi)
    removed: interval
def intervals_after_removal(intervals, removed): -> return sorted list of disjoint intervals

if len(intervals) == 0 -> return []

# Example

1------<3  4------<6
    2---------<5

1--<2          5--<6


1------<3  4----<6 7---------<9  10----<12
    2-------------------<8

1--<2                    8---<9  10----<12


# Algorithm

filtered = []
for each interval,
    compare with removed.
    case 1) no append.
    case 2) append part of interval
    case 3) append whole interval
return filtered

time  = O(N)
space = O(1)
'''

# Impl

# [1,2],[4,8],[10,14]]
# [3,11]
def filtered_intervals(intervals, removed):
    filtered = [] # [1,2],x,[11,14]

    for interval in intervals:# [1,2],[4,8],*[10,14]
        curr_compl = complement_interval(interval, removed)
        filtered.extend(curr_compl)

    return filtered

# Return array of complement intervals of i1 - i2
def complement_interval(i1, i2):
    start1, end1 = i1 # 1,4
    start2, end2 = i2 # 0,4

    if end1 <= start2 or end2 <= start1:
        return [i1]

    compls = []

    if start1 < start2:
        compls.append([start1, start2])
    if end2 < end1:
        compls.append([end2, end1])

    return compls


# 1------------<5
#     2---<3
        
# 1--<2    3---<5

    

# Test
assert filtered_intervals([[1,2],[4,8],[10,14]], [3,11]) == [[1,2],[11,14]]

assert filtered_intervals([[1,3]],[0,4]) == []      # complete
assert filtered_intervals([[1,3]],[1,4]) == []      # complete, same start
assert filtered_intervals([[1,4]],[0,4]) == []      # complete, same end
assert filtered_intervals([[1,3]],[2,4]) == [[1,2]] # parital 
assert filtered_intervals([[1,3]],[4,5]) == [[1,3]] # no overlap

assert filtered_intervals([],[1,2]) == [] # pass
```
