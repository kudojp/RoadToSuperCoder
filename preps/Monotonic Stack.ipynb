```python
# Final Prices With a Special Discount in a Shop

# You are given an integer array prices where prices[i] is the price of the ith item in a shop.

# There is a special discount for items in the shop. If you buy the ith item, then you will receive a discount equivalent to prices[j] where j is the minimum index such that j > i and prices[j] <= prices[i]. Otherwise, you will not receive any discount at all.

# Return an integer array answer where answer[i] is the final price you will pay for the ith item of the shop, considering the special discount.


'''
# 1.
# price is positive.
def final_prices(prices) -> return array of ints

# prices = [3, 5, 3, 7, 8, 6]
# disc   = [-3,-3,x,-6, -6,x]
# final  = [0, 2, 3, 1, 2, 6]

# 2.
# 3. BF
time  = N**2
space = N

# 4.
# prices = [3, 5, 6, 4, 2]

# strictly inreasing monotonic stack
time  = N
space = N

# 5.
final_prices = prices.copy()
stack = []
for each price:
    while stack top >= price:
        pop out and apply curr price bargain. and store it to final_prices
    push (curr price, index)
return final_prices
'''

# 6.


def final_prices(prices):
    finals = prices.copy() # [2,3,1] -> [2,2,1]->[1,2,1]
    stack = [] # (price, idx)     # x(2,0),x(3,1),(1,2)
 
    for curr_idx, curr_price in enumerate(prices): # 0:2, 1:3, 2:1*
        while 0 < len(stack) and stack[-1][0] >= curr_price:
            prev_price, prev_idx = stack.pop() # (3,1)->(2,0)
            finals[prev_idx] = prev_price - curr_price
        stack.append((curr_price, curr_idx))
    
    return finals


# 7.
assert final_prices([2,3,1]) == [1,2,1] # pass
assert final_prices([1,2,3]) == [1,2,3]
assert final_prices([3,2,1]) == [1,1,1]


assert final_prices([1]) == [1] # pass
assert final_prices([]) == []   # pass
```

```python
# Maximum Width Ramp
# A ramp in an integer array nums is a pair (i, j) for which i < j and nums[i] <= nums[j]. The width of such a ramp is j - i.

# Given an integer array nums, return the maximum width of a ramp in nums. If there is no ramp in nums, return 0.


'''
# 1.

if len(nums) <= 1 -> return None
def max_ramp_width(nums)


# 2.

1, 3, 2, 5, 0
<--->
<------>
<-------->     ----> return 4
   <----->
      <-->

# 3.

BF
time  = N**2
space = N**2

# 4.

1 2 3 4 5
1---------> return 4

5 4 3 2  x3  x4   5    6  4   6
         (3) (5) (7)  (x)

monotonic stack (+ hashmap)
time  = N * logN
space = N

# 5.
stack = [] # monotinic strictly decreasing stack
for each num:
    if num < stack top, push to stack.
    if not, while stack top <= num, get max ramp width which ends with num. and update global max.
return global max.
'''


# 6.
def max_ramp_width(nums): # 10, 8, 4, 9 -> 2
    if len(nums) <= 1:
        return None
    
    global_max = 0
    stack = [] # (0:10),(1,8),(2,4)

    for curr_idx, curr_num in enumerate(nums): # (3,9)
        # print("curr_num = ", curr_num)
        if len(stack) == 0 or curr_num < stack[-1][1]:
            stack.append((curr_idx, curr_num))
        else:
            # find index of first element which is smaller or equal to curr_num
            #  10, 8, 4, 4or8or12
            left, right = 0, len(stack)-1 # 0, 2
            # print("started with ", left, right)
            while left < right:
                center = (left + right) // 2 # 1
                if curr_num < stack[center][1]: # 9 <= 8 no.
                    left = center + 1
                else:
                    right = center # 2
            # print("ended with ", left, right)
            assert left == right
            prev_idx, _ = stack[left]

            local_max = curr_idx - prev_idx
            global_max = max(global_max, local_max)

    return global_max



# 10, 8, 4, 4
# 10, 8, 4, 9
# 10, 8, 4, 10
# 10, 8, 4, 12


# 7.
assert max_ramp_width([5,3,1,2,4]) == 3

assert max_ramp_width([]) == None
assert max_ramp_width([1]) == None
assert max_ramp_width([1,5]) == 1 # pass
assert max_ramp_width([1,1]) == 1 # pass
assert max_ramp_width([5,1]) == 0 # 0
```

```python
# Next Greater Node In Linked List
# You are given the head of a linked list with n nodes.

# For each node in the list, find the value of the next greater node.
# That is, for each node, find the value of the first node that is next to it and has a strictly larger value than it.

# Return an integer array answer where answer[i] is the value of the next greater node of the ith node (1-indexed).
# If the ith node does not have a next greater node, set answer[i] = 0.

'''
# 1.
def indexes_next_greter(head) -> return array of ints
- if head is None -> return []


# 2.
  
head -> 1 -> 5 -> 4 -> 3 -> 6
    [   5,   6    6    6    0]

# 3. BF
time  = n ** 2
space = 1

# 4.
mononically strictly decreasing stack.

head -> 1 -> 5 -> 4 -> 3 -> 6
(0,1)
(1,5)
stack (idx, values)
[x(0, 1)
returned array
[0->5,

# 5
for each (idx, val):
    append 0 to ret.
    while stack top < curr val,
        pop element from stack.
        fill in idx of ret with curr val
    push (idx, val) to stack

time = O(N)
space = O(N)
'''

# 6
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None

def indexes_next_greater(head): # 1->0->3
    if head is None:
        return []

    indexes = [0]             # 0,0,0 -> 0,3,0 -> 3,3,0
    stack = [(0, head.val)]   # x(0,1),x(1,0),(2,3)

    curr = head    # 1->0->3
    curr_idx = 0   # 0->1->2
    while curr.next is not None:
        curr = curr.next
        curr_idx += 1

        indexes.append(0)

        while 0 < len(stack) and stack[-1][1] < curr.val:
            prev_idx, _prev_val = stack.pop()
            indexes[prev_idx] = curr.val

        stack.append((curr_idx, curr.val))

    return indexes  

# 7
assert indexes_next_greater(None) == [] # pass

# increasing
l = Node(1)
l.next = Node(2)
l.next.next = Node(2)
l.next.next.next = Node(3)
assert indexes_next_greater(l) == [2,3,3,0]

# decreasing
l = Node(3)
l.next = Node(2)
l.next.next = Node(2)
l.next.next.next = Node(1)
assert indexes_next_greater(l) == [0,0,0,0]

# next greater is the second next one of current node.
l = Node(1)
l.next = Node(0)
l.next.next = Node(3)
assert indexes_next_greater(l) == [3,3,0]
```
