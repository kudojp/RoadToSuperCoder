```python
# Zigzag Conversion (33min)

# The string "PAYPALISHIRING" is written in a zigzag pattern on a given number of rows like this:
# (you may want to display this pattern in a fixed font for better legibility)

# P   A   H   N
# A P L S I I G
# Y   I   R

# And then read line by line: "PAHN APLSIIG YIR"

# Write the code that will take a string and make this conversion given a number of rows:

# string convert(string s, int numRows);


'''
# interface
string convert(string s, int numRows);


# "a", 3

a   x
x x x
x   x

-> return "a"


# examples

abcdefghijk,   l = 3

a   e   i
b d f h j
c   g   k

----> return "aei bdfhj cgk"

abcdefghijk,   l = 4


a |  g
b  f    
c e
d

(k * 2 - 2) chars for 1 pattern

# algorithm

[] <- idx % 6 == 0
[] <- idx % 6 == 1  , 5
[] <- idx % 6 == 2  , 4
[] <- idx % 6 == 3


time  = N + (N + k - 1) = N 
space = N + k - 1 = N

l = 3 -> 4
l = 4 -> 6
l = 5 -> 8


k = 5

0
1 7
2 6(6%5=1 and 3-1) <---- k-2 - modulo % k
3 5(5%5=0 and 3-0) 
4



# overview

- initaite k arryas
- for each number
    check idx % (k * 2 - 2)
    push this into appropriate array
- join them
'''
# impl
# 0    4
# 1  3 
# 2

# impl
def convert(s, k): # "abcdefg", 3
    if len(s) <= k:
        return "".join(s)
    
    # I forgot this !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!1
    if k == 1:
        return s

    rows = [[] for _ in range(k)]# [ae][b,d,][c]

    for idx, char in enumerate(s): # 0a, 1b, 2c, 3d, *4e, 5f, 6g
        modulo = idx % (k * 2 - 2) # idx % 4 -> 0, 1, 2, 3%4=3, 4%4=0

        if modulo < k: # m < 3
            row_idx = modulo
        else:
            row_idx = k-2 - modulo % k # 3-2 - 3%3 = 1
        rows[row_idx].append(char)
    
    row_strings = ["".join(row) for row in rows]#  [ae][b,d,f][c,g] -> ae, bdf, cg
    return "".join(row_strings)
        
# test
assert convert("abcdefg", 3) == "aebdfcg" # pass

assert convert("ab", 1) == "ab" # fail

assert convert("abc", 3) == "abc" # pass
assert convert("ab", 3) == "ab" # pass
assert convert("a", 3) == "a" # pass
assert convert("", 3) == "" # pass
```
