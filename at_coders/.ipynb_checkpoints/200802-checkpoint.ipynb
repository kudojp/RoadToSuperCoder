```python
n, d = map(lambda x: int(x), input().split())

count = 0
for _ in range(n):
    x, y = map(lambda x: int(x), input().split())
    if x**2 + y**2 <= d**2:
        count += 1
print(count)
```

```python
import sys
k = int(input())

remainders = set()
current_remainder = 7 % k
remainders.add(current_remainder)

counter = 1
while True:
    print(remainders, counter)
    counter += 1
    remainder = current_remainder * 10 % k
    if remainder in remainders:
        print('-1')
        sys.exit()
    current_remainder = current_remainder * 10 % k
    remainders.add(current_remainder)
    if sum(remainders) % k == 0:
        print(counter)
        sys.exit()
```

```python
import sys
k = int(input())

remainders = set()
remainder = 7 % k
remainders.add(remainder)

counter = 1
while True:
    counter += 1
    remainder = remainder * 10 % k
    if remainder in remainders:
        print('-1')
        sys.exit()
    remainders.add(remainder)
    if sum(remainders) % k == 0:
        print(counter)
        sys.exit()
```

```python
# c = [1,1]

import sys
def first_white(c):
    for i,ele in enumerate(c):
        if ele == 0:
            return i
    return -1
        
def last_red(c):
    for i in range(len(c)-1, -1, -1):
        if c[i] == 1:
            return i
    return -1
        
first_white(c), last_red(c)
```

```python
import sys

n = int(input())
c = [0 if x=='W' else 1 for x in input()]
counter = 0
while True:
    first_w = first_white(c)
    last_r = last_red(c)
    if (len(c) == 0) or (first_w < 0) or (last_r < 0) or (last_r < first_w):
        print(counter)
        sys.exit()
    c = c[first_w+1:last_r]    
    counter += 1
#     print(counter, c)
```

```python
import sys

n = int(input())
c = [0 if x=='W' else 1 for x in input()]
counter = 0

white_index = [i for i,ele in enumerate(c) if ele == 0]
red_index = [i for i in range(len(c)-1, -1, -1) if c[i] == 1]
    
    
for i in range(min(len(white_index), len(red_index))):
    if white_index[i] < red_index[i]:
        counter += 1
print(counter)
```
