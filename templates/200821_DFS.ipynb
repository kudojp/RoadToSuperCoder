# [A - 深さ優先探索 ](https://atcoder.jp/contests/atc001/tasks/dfs_a)


高橋君の住む街は長方形の形をしており、格子状の区画に区切られています。 長方形の各辺は東西及び南北に並行です。 各区画は道または塀のどちらかであり、高橋君は道を東西南北に移動できますが斜めには移動できません。 また、塀の区画は通ることができません。

高橋君が、塀を壊したりすることなく道を通って魚屋にたどり着けるかどうか判定してください。

入力入力は以下の形式で標準入力から与えられる。

1 行目には、街の南北の長さとして整数 H(1≦H≦500) と東西の長さとして整数 W(1≦W≦500) が空白で区切られて与えられる。2 行目からの H 行には、格子状の街の各区画における状態 ci,j(0≦i≦H−1,0≦j≦W−1) が与えられる。i 行目 j 文字目の文字 ci,j はそれぞれ s, g, ., # のいずれかで与えられ、座標 (j,i) が下記のような状態であることを表す。
- `s` : その区画が家であることを表す。
- `g` : その区画が魚屋であることを表す。
- `.` : その区画が道であることを表す。
- `#` : その区画が塀であることを表す。

高橋君は家・魚屋・道は通ることができるが、塀は通ることができない。与えられた街の外を通ることはできない。s と g はそれぞれ 1 つずつ与えられる。出力塀を 1 回も壊さずに、家から魚屋まで辿り着くことができる場合は Yes、辿りつけない場合は No を標準出力に 1 行で出力せよ。

## スタックを使う

````python
import sys

def next_visitable_indices(current_index, unvisited_indices, n_col):
    for i in [-n_col, n_col]:
        indices = set([current_index + i for i in [-n_col, n_col] if (current_index + i) in unvisited_indices])
    #　cell not on the right edge
    if (current_index + 1) % n_col != 0:
        if (current_index + 1) in unvisited_indices:
            indices.add(current_index+1) 
    #　cell not on the left edge
    if current_index % n_col != 0:
        if (current_index + -1) in unvisited_indices:
            indices.add(current_index-1) 
    return indices
        
    

h, w = map(lambda x: int(x), input().split())

unvisited_indices = set()

for i_row in range(h):
    for i_col, c in enumerate(input()):
        index = w * i_row + i_col
        if c == 's':
            start_index = index
        if c == 'g':
            goal_index = index
            unvisited_indices.add(index)
        if c == '.':
            unvisited_indices.add(index)
# print(start_index, goal_index, unvisited_indices)


current_index = start_index
stack = []

while True:
    next_indices = next_visitable_indices(current_index, unvisited_indices, w)
#     print(current_index, next_indices)
    
    if goal_index in next_indices:
        print("Yes")
        sys.exit()
    
    while len(next_indices) == 0:
        # step back
        try:
            current_index = stack.pop()
            next_indices = next_visitable_indices(current_index, unvisited_indices, w)  
#             print(current_index, next_indices)
        except IndexError:
            print('No')
            sys.exit()

    next_index = next_indices.pop()
    unvisited_indices.remove(next_index)
    stack.append(current_index)
    
    # step to next_index
    current_index = next_index
    #  print(stack)
````

### 以下の例でwaになっていた。
理由は、`def next_visitable_indices`によって、右端(or左端)にあるcellに+1(-1)したセルを含んだセットが返されていたから。

以下の例だと、index=9からindex=10に移動できるような実装になっていた。
